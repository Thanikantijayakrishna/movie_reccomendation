```python
import pandas as pd
```

```python
chunks = []
for chunk in pd.read_csv(
        r'C:/Users/hi/Desktop/VI/AI/mlprojects/mymoviedb.csv',
        engine='python',
        sep=',',
        chunksize=100_000,   # tune chunk size as needed
        on_bad_lines='skip'):
    chunks.append(chunk)

df = pd.concat(chunks, ignore_index=True)
df.head()
```

```python
df.info()
# df.describe()
```

```python
def combine_features(row):
    return f"{row['Title']} {row['Overview']} {row['Genre']}"

df['combined_features'] = df.apply(combine_features, axis=1)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(stop_words='english')
tfidf_matrix = tfidf.fit_transform(df['combined_features'])
```

```python
print(tfidf_matrix)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
# Reset index to safely use indices for lookup
df = df.reset_index()

indices = pd.Series(df.index, index=df['Title'].str.lower())

def recommend_movies(title, cosine_sim=cosine_sim):
    title = title.lower()
    if title not in indices:
        return "Movie not found."

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:50]  # top 10 similar movies

    movie_indices = [i[0] for i in sim_scores]
    return df['Title'].iloc[movie_indices]

```

```python
import streamlit as st
import pandas as pd
import joblib
import csv
import sys

# ──────────────────────────────────────────────────────────────
# 1. Load Data & Artifacts
# ──────────────────────────────────────────────────────────────
@st.cache_resource(show_spinner="Loading data and similarity matrix…")
def load_artifacts(csv_path: str):
    # Fix OverflowError
    csv.field_size_limit(2**20)

    # Load CSV in chunks to handle large files
    chunks = []
    for chunk in pd.read_csv(
            r'C:/Users/hi/Desktop/VI/AI/mlprojects/mymoviedb.csv',
            engine='python',
            sep=',',
            chunksize=100_000,   # tune chunk size as needed
            on_bad_lines='skip'):
        chunks.append(chunk)

    df = pd.concat(chunks, ignore_index=True)
    # df.head()

    # Clean Title column
    df['Title'] = df['Title'].astype(str)

    # Load precomputed similarity objects
    tfidf_matrix = joblib.load('tfidf_matrix.pkl')
    cosine_sim = joblib.load('cosine_sim.pkl')
    indices = joblib.load('indices.pkl')

    # Make index keys lowercase for consistent lookup
    indices.index = indices.index.str.lower()

    return df, cosine_sim, indices

df, cosine_sim, indices = load_artifacts("mymoviedb.csv")

# ──────────────────────────────────────────────────────────────
# 2. Recommendation logic
# ──────────────────────────────────────────────────────────────
def recommend_movies(title: str, df: pd.DataFrame, cosine_sim, indices, top_n: int = 10) -> pd.DataFrame:
    title = title.lower()
    if title not in indices:
        return pd.DataFrame()  # empty result

    idx = indices[title]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)[1:top_n+1]
    movie_indices = [i[0] for i in sim_scores]

    results = df[['Title', 'Poster_Url']].iloc[movie_indices]
    return results.reset_index(drop=True)

# ──────────────────────────────────────────────────────────────
# 3. Streamlit UI
# ──────────────────────────────────────────────────────────────
st.set_page_config(page_title="🎬 Movie Recommender", layout="wide")
st.title("🎬 Movie Recommendation System")
st.write("Type a movie title to get similar movie recommendations!")

movie_input = st.text_input("Enter a movie title", "")

if movie_input:
    results = recommend_movies(movie_input, df, cosine_sim, indices)

    if not results.empty:
        st.success("Top recommended movies:")
        cols = st.columns(2)

        for i, row in results.iterrows():
            with cols[i % 2]:
                st.image(row['Poster_Url'], width=150)
                st.markdown(f"**{row['Title']}**")
    else:
        st.error("Sorry, no similar movies found or movie not in database.")
```
